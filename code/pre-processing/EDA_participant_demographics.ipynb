```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Participant demographic data 

```python
df_demographic = pd.read_excel('../participant_demographics_table.xlsx')
df_demographic
```

```python
# Define the stage order
stage_order = ['One', 'Two', 'Three', 'Four']
```

#### 1. Sex VS Hoehn and Yahr (stage)

```python
# Sex VS Hoehn and Yahr (stage)
sns.countplot(data=df_demographic, x='Hoehn and Yahr (stage)', hue='Sex', palette='pastel', order=stage_order)
plt.title('Gender Distribution Across Disease Stages')
plt.xlabel('Hoehn and Yahr Stage')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

#### 2. Hoehn and Yahr (stage) VS Years since diagnosis

```python
# Hoehn and Yahr (stage) VS Years since diagnosis

sns.boxplot(data=df_demographic,  x='Hoehn and Yahr (stage)',  y='Years since diagnosis', 
    hue='Hoehn and Yahr (stage)', palette='pastel',  order=stage_order)
plt.title('Disease Stage vs. Years Since Diagnosis')
plt.xlabel('Hoehn and Yahr Stage')
plt.ylabel('Years Since Diagnosis')
plt.show()
```

#### 3. Age VS Years since diagnosis

```python
# Age VS Years since diagnosis
sns.regplot(data=df_demographic, x='Age', y='Years since diagnosis', scatter_kws={'s': 30})
plt.title('Age vs. Years Since Diagnosis')
plt.xlabel('Age (years)')
plt.ylabel('Years Since Diagnosis')
```

#### 4. Hoehn and Yahr (stage) VS UPDRS

```python
# Hoehn and Yahr (stage) VS UPDRS：
sns.boxplot(data=df_demographic, x='Hoehn and Yahr (stage)', y='UPDRS', 
    hue='Hoehn and Yahr (stage)', palette='pastel', order=stage_order)
plt.title('Disease Stage vs. UPDRS Score')
plt.xlabel('Hoehn and Yahr Stage')
plt.ylabel('UPDRS Score')
plt.show()
```

#### 5. Sex VS UPDRS

```python
# Sex VS UPDRS
sns.boxplot(data=df_demographic, x='Sex', y='UPDRS', palette='pastel', hue='Sex')
plt.title('Gender vs. UPDRS Score')
plt.xlabel('Gender')
plt.ylabel('UPDRS Score')
```

#### 6. Age VS UPDRS

```python
# Age VS UPDRS
sns.scatterplot(data=df_demographic, x='Age', y='UPDRS', hue='Sex', s=50)
plt.title('Age vs. UPDRS Score')
plt.xlabel('Age (years)')
plt.ylabel('UPDRS Score')
plt.show()
```

#### 7. Sex VS CIT

```python
# Sex VS CIT
sns.boxplot(data=df_demographic, x='Sex', y='CIT', hue='Sex' , palette='pastel')
plt.title('Gender vs. CIT Score')
plt.xlabel('Gender')
plt.ylabel('CIT Score')
plt.show()
```

#### 8. Age VS CIT

```python
# Age VS CIT
sns.regplot(data=df_demographic, x='Age', y='CIT', scatter_kws={'s': 20})
plt.title('Age vs. CIT Score')
plt.xlabel('Age (years)')
plt.ylabel('CIT Score')
```

#### 9. CIT vs UPDRS

```python
# CIT vs UPDRS
sns.scatterplot(data=df_demographic, x='CIT', y='UPDRS', hue='Sex', s=100, palette='pastel')
plt.title('CIT Score vs. UPDRS Score')
plt.xlabel('CIT Score')
plt.ylabel('UPDRS Score')
plt.show()
```

#### 10. CIT VS Years since diagnosis

```python
# CIT VS Years since diagnosis
sns.regplot(data=df_demographic, x='Years since diagnosis', y='CIT', scatter_kws={'s': 20})
plt.title('Years Since Diagnosis vs. CIT Score')
plt.xlabel('Years Since Diagnosis')
plt.ylabel('CIT Score')
```

#### 11. Hoehn and Yahr (stage) VS Dominant Side

```python
# Hoehn and Yahr (stage) VS Dominant Side
sns.countplot(data=df_demographic, x='Hoehn and Yahr (stage)', hue='Dominant Side', 
              palette='pastel', order=stage_order)
plt.title('Disease Stage vs. Dominant Side')
plt.xlabel('Hoehn and Yahr Stage')
plt.ylabel('Count')
plt.legend(title='Dominant Side', loc='upper right')
plt.show()
```
